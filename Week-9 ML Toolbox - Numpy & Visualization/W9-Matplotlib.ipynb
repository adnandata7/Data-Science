#### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.
<hr>

A scatter plot is a type of data visualization that uses Cartesian coordinates to display values typically for two variables. Each point on the scatter plot represents an observation in the dataset. The position of the point is determined by the values of the two variables: one variable is plotted along the x-axis and the other along the y-axis. Scatter plots are particularly useful for showing the relationship or correlation between two variables.

```python
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel("X data")
plt.ylabel("Y data")
plt.title("This is scatter plot")
```

#### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and  y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
<hr>

The subplot() function in Matplotlib is used to create multiple plots (subplots) in a single figure. This is useful when you want to compare different plots side by side or when you have limited space and want to organize your visualizations efficiently.

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])


plt.figure(figsize=(10, 8))

# First subplot
plt.subplot(2, 2, 1)  
plt.plot(x1, y1, label="Line 1")
plt.title("Line 1")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Second subplot
plt.subplot(2, 2, 2)
plt.plot(x2, y2, label="Line 2")
plt.title("Line 2")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Third subplot
plt.subplot(2, 2, 3)
plt.plot(x3, y3, label="Line 3")
plt.title("Line 3")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Fourth subplot
plt.subplot(2, 2, 4)
plt.plot(x4, y4, label="Line 4")
plt.title("Line 4")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Adjust layout
plt.tight_layout()
```

#### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
<hr>

A bar plot (or bar chart) is a type of visualization used to represent data with rectangular bars. The length or height of each bar corresponds to the value it represents, allowing for easy comparison across different categories or groups. Bar plots can be displayed vertically or horizontally.

Bar plot is used for:  
Comparison: Bar plots are excellent for comparing quantities across different categories. They make it easy to see which category is the largest or smallest.  
Clarity: They provide a clear and straightforward way to present discrete data. The visual separation of bars helps in distinguishing between different categories.  
Categorical Data: They are ideal for visualizing categorical data, where each bar represents a category or a group.  
Trends Over Time: They can show trends over time if the categories represent different time periods.  
Versatility: Bar plots can be used for a wide range of data types and can include both positive and negative values.  

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
plt.plot(company, profit )
plt.title("Bar graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)
<hr>

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays the dataset's central tendency, variability, and potential outliers. The plot is named for its "box" shape and the "whiskers" that extend from the box to indicate the range of the data.

Box plot is used for:   
Summarize Data Distribution: Provides a summary of the data distribution, highlighting the central tendency, spread, and skewness.  
Identify Outliers: Easily identifies outliers in the data, which can be important for data analysis and decision-making.  
Compare Multiple Datasets: Facilitates comparison of distributions across multiple datasets or groups.  
Highlight Data Symmetry: Shows whether the data is symmetrically distributed or skewed.  
Handle Large Datasets: Efficiently summarizes large datasets without displaying all individual data points.  

```python
box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1, box2] , vert=True, patch_artist=True)
plt.title("Box Plot Example")
plt.xlabel("Groups")
plt.ylabel("Values")
```
