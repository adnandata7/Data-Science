### Q1: Explain the following with an example:

1. Artificial Intelligence (AI)

Definition: AI is the simulation of human intelligence in machines that are programmed to think and learn like humans. It encompasses various fields such as machine learning, natural language processing, robotics, and computer vision.  
Example: Virtual assistants like Siri or Alexa use AI to understand and respond to user queries by processing natural language, retrieving information, and executing tasks.

2. Machine Learning (ML)

Definition: ML is a subset of AI that focuses on the development of algorithms and statistical models that enable computers to learn from and make predictions or decisions based on data.  
Example: A spam email filter that learns to identify and classify spam emails from non-spam emails based on patterns and features extracted from previously labeled emails.

3. Deep Learning (DL)

Definition: DL is a subset of ML that uses neural networks with many layers (deep neural networks) to model complex patterns in large datasets. It is particularly effective in tasks such as image and speech recognition.  
Example: An image recognition system that can accurately classify objects in photographs, such as identifying animals in wildlife photos, by training deep convolutional neural networks (CNNs) on a large dataset of labeled images.

### Q2: What is supervised learning? List some examples of supervised learning.

Supervised Learning:

Definition: Supervised learning is a type of machine learning where the model is trained on a labeled dataset, meaning that each training example is paired with an output label. The model learns to map inputs to the correct output.  
Examples:  
Linear Regression: Predicting house prices based on features like size, location, and number of rooms.  
Logistic Regression: Classifying emails as spam or non-spam.  
Decision Trees: Predicting whether a customer will buy a product based on their demographic information.  
Support Vector Machines (SVM): Classifying images of cats and dogs.  
Neural Networks: Recognizing handwritten digits in images.

### Q3: What is unsupervised learning? List some examples of unsupervised learning.

Unsupervised Learning:

Definition: Unsupervised learning is a type of machine learning where the model is trained on data without labeled responses. The goal is to infer the natural structure present within a set of data points.  
Examples:  
Clustering: Grouping customers based on purchasing behavior (e.g., using K-means clustering).  
Dimensionality Reduction: Reducing the number of features in a dataset while retaining important information (e.g., using Principal Component Analysis (PCA)).  
Anomaly Detection: Identifying unusual patterns that do not conform to expected behavior (e.g., detecting fraudulent transactions).

### Q4: What is the difference between AI, ML, DL, and DS?

Artificial Intelligence (AI): The broadest field that includes any technique enabling machines to mimic human behavior.

Machine Learning (ML): A subset of AI focused on algorithms that learn from data to make predictions or decisions.

Deep Learning (DL): A subset of ML that uses deep neural networks to model complex patterns in large datasets.

Data Science (DS): An interdisciplinary field that uses scientific methods, processes, algorithms, and systems to extract knowledge and insights from structured and unstructured data. It often incorporates AI, ML, and DL as tools for data analysis and prediction.

### Q5: What are the main differences between supervised, unsupervised, and semi-supervised learning?

Supervised Learning:

Data: Labeled data.  
Goal: Learn a mapping from inputs to outputs.  
Examples: Classification, regression.  

Unsupervised Learning:

Data: Unlabeled data.  
Goal: Discover hidden patterns or intrinsic structures in input data.  
Examples: Clustering, dimensionality reduction.

Semi-Supervised Learning:

Data: A small amount of labeled data and a large amount of unlabeled data.  
Goal: Improve learning accuracy by using both labeled and unlabeled data.  
Examples: Classification with limited labeled examples.  

### Q6: What is train, test, and validation split? Explain the importance of each term.

Train Set: The subset of data used to train the model. The model learns patterns and relationships from this data. 

Test Set: The subset of data used to evaluate the performance of the trained model. It provides an unbiased assessment of the model's accuracy and generalization to new data.

Validation Set: The subset of data used to tune model parameters and make decisions about the model architecture. It helps in selecting the best model and avoiding overfitting.

Importance:

Train Set: Ensures the model learns effectively.  
Test Set: Ensures the model generalizes well to unseen data.  
Validation Set: Helps in model selection and parameter tuning, preventing overfitting.

### Q7: How can unsupervised learning be used in anomaly detection?

Unsupervised learning can be used in anomaly detection by identifying patterns and structures in the data and then detecting deviations from these patterns. Techniques such as clustering can group normal data points together, making it easier to spot outliers. Principal Component Analysis (PCA) can reduce data dimensions, highlighting anomalies that deviate significantly from the main components.
