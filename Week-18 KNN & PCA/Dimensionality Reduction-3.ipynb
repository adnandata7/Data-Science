### Q1. What are Eigenvalues and Eigenvectors? How are they related to the Eigen-Decomposition approach? Explain with an example.

**Eigenvalues** and **eigenvectors** are fundamental in linear algebra. Given a square matrix \( A \), a non-zero vector \( \vec{v} \) is called an **eigenvector** of \( A \) if it satisfies:

$$
A\vec{v} = \lambda\vec{v}
$$

Here, \( \lambda \) is the **eigenvalue** corresponding to the eigenvector \( \vec{v} \). This equation means that when \( A \) acts on \( \vec{v} \), the result is just a scaled version of \( \vec{v} \).

**Example**:

Let

$$
A = \begin{bmatrix} 2 & 0 \\ 0 & 3 \end{bmatrix}
$$

The eigenvalues are 2 and 3, with eigenvectors:

$$
\vec{v}_1 = \begin{bmatrix} 1 \\ 0 \end{bmatrix}, \quad \vec{v}_2 = \begin{bmatrix} 0 \\ 1 \end{bmatrix}
$$

**Eigen-Decomposition** expresses a matrix as:

$$
A = PDP^{-1}
$$

Where:
- \( D \) is a diagonal matrix of eigenvalues
- \( P \) is a matrix whose columns are eigenvectors of \( A \)

### Q2. What is eigen decomposition and what is its significance in linear algebra?

**Eigen decomposition** is the process of decomposing a square matrix into its eigenvalues and eigenvectors:

$$
A = PDP^{-1}
$$

It is significant because:
- It simplifies matrix operations like powers and exponentials.
- It provides insight into the matrixâ€™s structure.

### Q3. What are the conditions that must be satisfied for a square matrix to be diagonalizable using the Eigen-Decomposition approach? Provide a brief proof to support your answer.

A square matrix \( A \) is **diagonalizable** if:
- It has **n linearly independent eigenvectors**, where \( n \) is the size of the matrix.

This occurs if:
- \( A \) has \( n \) **distinct eigenvalues**, or
- The **algebraic multiplicity equals the geometric multiplicity** for each eigenvalue.

**Proof (Outline)**:
If \( A \) has \( n \) linearly independent eigenvectors \( \vec{v}_1, \vec{v}_2, \ldots, \vec{v}_n \), then:

$$
A = PDP^{-1}
$$

Where \( P \) contains \( \vec{v}_i \) as columns, and \( D \) has the corresponding \( \lambda_i \) on the diagonal.

### Q4. What is the significance of the spectral theorem in the context of the Eigen-Decomposition approach? How is it related to the diagonalizability of a matrix? Explain with an example.

The **spectral theorem** states that:

> Any real symmetric matrix is **orthogonally diagonalizable**, i.e.,

$$
A = QDQ^T
$$

Where:
- \( Q \) is an orthogonal matrix (\( Q^T = Q^{-1} \))
- \( D \) is a diagonal matrix of real eigenvalues

**Significance**:
- Guarantees diagonalization for symmetric matrices
- Ensures numerical stability

**Example**:

Let

$$
A = \begin{bmatrix} 4 & 1 \\ 1 & 3 \end{bmatrix}
$$

This is symmetric and can be diagonalized via the spectral theorem.

### Q5. How do you find the eigenvalues of a matrix and what do they represent?

To find eigenvalues of matrix \( A \), solve the **characteristic equation**:

$$
\det(A - \lambda I) = 0
$$

The solutions \( \lambda \) are the eigenvalues.

**Interpretation**: They represent the **scaling factors** by which the matrix transforms its eigenvectors.

---

### Q6. What are eigenvectors and how are they related to eigenvalues?

**Eigenvectors** are non-zero vectors that **do not change direction** under a linear transformation, only scaled by a factor.

Given:

$$
A\vec{v} = \lambda\vec{v}
$$

- \( \vec{v} \) is an eigenvector
- \( \lambda \) is the corresponding eigenvalue

---

### Q7. Can you explain the geometric interpretation of eigenvectors and eigenvalues?

Geometrically:
- **Eigenvectors** point in directions that remain unchanged under the matrix transformation.
- **Eigenvalues** indicate how much the eigenvectors are **stretched or compressed**.

For example, a rotation matrix in 2D has no real eigenvectors, as no direction remains fixed.

### Q8. What are some real-world applications of eigen decomposition?

- **Principal Component Analysis (PCA)**: Reducing dimensionality by projecting data on principal eigenvectors.
- **Vibration Analysis**: Modes of vibration correspond to eigenvalues/eigenvectors.
- **Facial Recognition**: Eigenfaces technique uses eigen decomposition for feature extraction.

---

### Q9. Can a matrix have more than one set of eigenvectors and eigenvalues?

- **Eigenvalues** are unique (up to multiplicity).
- **Eigenvectors** corresponding to the same eigenvalue are not unique. Any scalar multiple of an eigenvector is also a valid eigenvector.
- A matrix can have multiple sets of eigenvectors if eigenvalues are repeated (degenerate).
