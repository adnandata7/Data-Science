### **Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.**

**(i). PMF**:

It is used for discrete random variables and gives the probability of each possible outcome.

It is a function that maps each possible value to its probability.

**(ii). PDF**:

It is used for continuous random variables and represents the likelihood of the variable taking on a specific value.

It gives the probability density at each point.

*Example*: For a fair six-sided die, the PMF assigns a probability of 1/6 to each face (1, 2, 3, 4, 5, 6).

In contrast, for a continuous uniform distribution between 0 and 1, the PDF is constant, and the area under the curve between any two points is proportional to the probability.

### **Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?**

**CDF**:

It gives the probability that a random variable takes on a value less than or equal to a specific value. 

It accumulates probabilities as you move along the distribution.

*Example*: The CDF of a fair six-sided die would show that the probability of rolling a 3 or less is 1/2 (because there are 3 favorable outcomes out of 6 possible).

### **Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset is normally distributed, what is the probability that a randomly selected observation will be greater than 60? Use the appropriate formula and show your calculations.**

**The formula to calculate the z-score is:**

$$Z = \frac{X - \mu}{\sigma}$$

Where:

- $X$ is the value you want to find the probability for (in this case, 60).
- $\mu$ is the mean of the dataset (50).
- $\sigma$ is the standard deviation of the dataset (10).

Calculating the z-score for $X=60$:

$$Z = \frac{60 - 50}{10} = 1$$

Using a standard normal distribution table, the probability associated with $Z=1$ is approximately 0.8413. This means that the probability of randomly selecting an observation greater than 60 from this normally distributed dataset is approximately **0.8413, i.e., 84.13%.**

### **Q8: What is the z score? State the importance of the z score.**

**Z-Score**:

It measures how many standard deviations a data point is from the mean.

**Importance**:

### **Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.**

**Central Limit Theorem**:

CLT states that the distribution of the sum or average of a large number of independent, identically distributed random variables approaches a normal distribution.

**Significance**:

It allows us to make inferences about a population by studying samples, regardless of the population's underlying distribution.

---
### **Q10: State the assumptions of the Central Limit Theorem.**

**Assumptions of the Central Limit Theorem**:

(i). Samples must be random and independent.

(ii). Samples must be sufficiently large (typically n ≥ 30).

(iii). The underlying population should have a finite mean and variance.
