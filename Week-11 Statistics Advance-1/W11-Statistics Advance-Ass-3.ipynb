### **Q1: What is Estimation Statistics? Explain point estimate and interval estimate.**

**Estimation Statistics** is a branch of statistics that deals with estimating population parameters based on sample data. 

**There are two types of estimation:**

- **Point Estimate**: A point estimate is a single value that is used to estimate a population parameter. 

    It is typically calculated as a sample statistic, such as the sample mean or sample proportion.

- **Interval Estimate**: An interval estimate provides a range of values within which the population parameter is likely to fall. 

    It is expressed as a confidence interval, which consists of a lower and upper limit based on the sample data.

### **Q2. Write a Python function to estimate the population mean using a sample mean and standard deviation.**

```python
import math

def estimate_population_mean(sample_mean, sample_std, sample_size, confidence_level=0.95):
    z_critical = 1.96 
    margin_of_error = z_critical * (sample_std / math.sqrt(sample_size))
    
    lower_limit = sample_mean - margin_of_error
    upper_limit = sample_mean + margin_of_error
    
    return lower_limit, upper_limit

# Example usage:
sample_mean = 50 
sample_std = 10  
sample_size = 100  

lower, upper = estimate_population_mean(sample_mean, sample_std, sample_size)
print(f"Estimated population mean: {sample_mean} (95% Confidence Interval: {lower} to {upper})")
```

### **Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.**

**Hypothesis testing** is a statistical method used to make inferences about population parameters based on sample data. 

It involves formulating a null hypothesis (H0) and an alternative hypothesis (Ha), collecting sample data, and using statistical tests to determine whether there is enough evidence to reject the null hypothesis in favor of the alternative hypothesis.

Hypothesis testing is used for several reasons:

- To make decisions and draw conclusions about population parameters.
- To assess the significance of an effect or relationship in a study.
- To evaluate the validity of a claim or statement.
- To test the effectiveness of a treatment or intervention.

The importance of hypothesis testing lies in its ability to provide a structured and objective way to make informed decisions and draw valid conclusions based on data.

### **Q5. Write a Python script to conduct a hypothesis test on the difference between two population means, given a sample from each population.**

### **Q6: What is a null and alternative hypothesis? Give some examples.**

- **Null Hypothesis (H0)**: The null hypothesis is a statement of no effect or no difference. It represents the status quo or the default assumption. It is what you aim to test against. Examples:
  - H0: The mean weight of product A is equal to the mean weight of product B.
  - H0: There is no difference in test scores between two groups.

- **Alternative Hypothesis (Ha)**: The alternative hypothesis is what you're trying to establish. It contradicts the null hypothesis and suggests the presence of an effect or difference. Examples:
  - Ha: The mean weight of product A is not equal to the mean weight of product B.
  - Ha: There is a significant difference in test scores between two groups.

### **Q7: Write down the steps involved in hypothesis testing.**

The steps involved in hypothesis testing are as follows:

1. **Formulate Hypotheses:**
   - Null Hypothesis (H0): Represents the status quo or no effect.
   - Alternative Hypothesis (Ha): Represents the effect you want to test for.

2. **Collect Data:**
   - Gather sample data relevant to the hypothesis being tested.

3. **Choose Significance Level (Alpha):**
   - Determine the desired level of significance (alpha), typically set at 0.05.

4. **Perform Test:**
   - Select an appropriate statistical test based on the type of data and research question.
   - Calculate the test statistic.

5. **Determine Critical Region:**
   - Define the critical region(s) in the distribution of the test statistic.

6. **Calculate P-Value:**
   - Find the probability of observing a test statistic as extreme as the one calculated, assuming the null hypothesis is true.

7. **Compare P-Value to Alpha:**
   - If the p-value is less than or equal to alpha, reject the null hypothesis.
   - If the p-value is greater than alpha, fail to reject the null hypothesis.

8. **Draw Conclusions:**
   - Based on the comparison of p-value and alpha, draw conclusions about the null hypothesis.

9. **Report Results:**
   - Communicate the results, including the test statistic, p-value, and whether the null hypothesis was rejected or not.

### **Q8. Define p-value and explain its significance in hypothesis testing.**

**P-Value (Probability Value)**: The p-value is a measure of the evidence against the null hypothesis. 

It quantifies the probability of obtaining a test statistic as extreme as, or more extreme than, the one calculated from the sample data, assuming that the null hypothesis is true.

In other words, it assesses the strength of evidence against the null hypothesis.

**Significance in Hypothesis Testing:**
- If the p-value is small (typically less than alpha), it suggests that the observed data provides strong evidence against the null hypothesis, leading to the rejection of the null hypothesis.
- If the p-value is large, it indicates that the observed data is consistent with the null hypothesis, leading to a failure to reject the null hypothesis.

In summary, 
    the p-value helps researchers make decisions about the null hypothesis. A smaller p-value suggests stronger evidence against the null hypothesis and supports the alternative hypothesis.

### **Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two random samples of equal size and a null hypothesis that the population means are equal.**

```python
import scipy.stats as stats
import numpy as np

sample1 = np.random.normal(50, 10, 100)  # Sample 1 (mean=50, std=10)
sample2 = np.random.normal(50, 10, 100)  # Sample 2 (mean=50, std=10)

# Perform a two-sample t-test
t_stat, p_value = stats.ttest_ind(sample1, sample2)
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in means.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in means.")

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
```

### **Q12: What is t-statistic? State the formula for t-statistic.**

The **t-statistic** is a measure used in hypothesis testing to assess the difference between a sample statistic and a population parameter. 

**It is calculated using the formula:**

$$t = \frac{\text{Sample Mean} - \text{Population Mean}}{\frac{\text{Sample Standard Deviation}}{\sqrt{\text{Sample Size}}}}$$

where:
- Sample Mean is the mean of the sample data.
- Population Mean is the hypothesized value of the population parameter (usually under the null hypothesis).
- Sample Standard Deviation is the standard deviation of the sample data.
- Sample Size is the number of observations in the sample.

### **Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.**

To test the hypothesis, you can use a one-sample t-test. The null and alternative hypotheses are:

- Null Hypothesis (H0): The new drug has no effect, i.e., the mean decrease in blood pressure is 10 mmHg (μ = 10).
- Alternative Hypothesis (Ha): The new drug decreases blood pressure by a different amount, i.e., the mean decrease in blood pressure is not equal to 10 mmHg (μ ≠ 10).

- **Sample Mean (x̄) = 8 mmHg**
- **Sample Standard Deviation (s) = 3 mmHg**
- **Sample Size (n) = 100**
- **Significance Level (alpha) = 0.05**

Calculate the t-statistic:

$$t = \frac{8 - 10}{\frac{3}{\sqrt{100}}}$$

$$t = \frac{-2}{0.3} = -6.67$$

Degrees of Freedom (df) = n - 1 = 100 - 1 = 99

Find the critical t-values for a two-tailed test at alpha = 0.05 and df = 99. You will find t-critical values of approximately -1.9845 and 1.9845.

Since the calculated t-statistic (-6.67) is much more extreme than the critical t-values, you reject the null hypothesis.

Therefore, you can conclude that the new drug has a significant effect on decreasing blood pressure.

### **Q17. A marketing company wants to estimate the average number of ads watched by viewers during a TV program. They take a random sample of 50 viewers and find that the sample mean is 4 with a standard deviation of 1.5. Estimate the population mean with a 99% confidence interval.**

To estimate the population mean with a 99% confidence interval, you can use the t-distribution. The formula for the confidence interval is:

$$ \text{Confidence Interval} = \text{Sample Mean} \pm \left(\frac{t \cdot \text{Sample Standard Deviation}}{\sqrt{\text{Sample Size}}}\right) $$

- Sample Mean $(\bar{x}): 4$

- Sample Size $(n)$: 50

- Standard Deviation $(\sigma)$: 1.5

- Degrees of Freedom (df) = n - 1 = 50 - 1 = 49

- Confidence Level: 99%

For a 99% confidence level, the Z-score is approximately 2.576.

Now, plug these values into the formula:

$$\text{Confidence Interval} = 4 \pm 2.576 \left( \frac{1.5}{\sqrt{50}} \right)$$

Now, calculate the values:

$$\text{Confidence Interval} = 4 \pm 2.576 \times 0.21213$$

Now, calculate the lower and upper bounds of the confidence interval:

**Lower Bound:** $4 - 2.576 \times 0.21213 = 4 - 0.54707 \approx 3.45293$

**Upper Bound:** $4 + 2.576 \times 0.21213 = 4 + 0.54707 \approx 4.54707$

So, the 99% confidence interval for the population mean number of ads watched by viewers during the TV program is approximately **3.45293** to **4.54707**.
